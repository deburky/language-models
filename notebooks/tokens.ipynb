Given a text string (e.g., *"tiktoken is great!"*) and an encoding (e.g., *"cl100k_base"*), a tokenizer can split the text string into a list of tokens (e.g., `["t", "ik", "token", " is", " great", "!"]`).

```python
from transformers import GPT2Tokenizer

# Load GPT-2 tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Example sentence
text = "unhappiness and hyperrealistic paintings"

# Encode text to tokens
tokens = tokenizer.encode(text)
print("Tokens:", tokens)

# Convert tokens back to text
decoded_text = tokenizer.decode(tokenizer.encode(text))
print("Decoded:", decoded_text)

# Tokenize the text
subword_tokens = tokenizer.tokenize(text)
print("Subword tokens:", subword_tokens)
```

```python
import tiktoken

# Load GPT-2 encoding (used in OpenAI's models)
tokenizer = tiktoken.get_encoding("gpt2")

text = "unhappiness and hyperrealistic paintings"

# Encode text (convert to token IDs)
tokens = tokenizer.encode(text)
print("Token IDs:", tokens)

# Decode tokens (convert back to text)
decoded_text = tokenizer.decode(tokens)
print("Decoded text:", decoded_text)

# Convert token IDs to readable subword tokens
subword_tokens = [tokenizer.decode([t]) for t in tokens]
print("Subword tokens:", subword_tokens)
```

## Spaces and subwords

In English, tokens commonly range in length from one character to one word (e.g., "t" or " great"), though in some languages tokens can be shorter than one character or longer than one word. Spaces are usually grouped with the starts of words (e.g., " is" instead of "is " or " "+"is"). You can quickly check how a string is tokenized at the [OpenAI Tokenizer](https://beta.openai.com/tokenizer), or the third-party [Tiktokenizer webapp](https://tiktokenizer.vercel.app/).

[**📑 How to count tokens with Tiktoken**](https://cookbook.openai.com/examples/how_to_count_tokens_with_tiktoken)

```python
import tiktoken

tokenizer = tiktoken.get_encoding("gpt2")

tokens = tokenizer.encode("unhappiness and hyperrealistic paintings")

# Decode as raw bytes
decoded_bytes = tokenizer.decode_tokens_bytes(tokens)
print(decoded_bytes)  # List of byte sequences
```

```python
import tiktoken

# Load the GPT-2 tokenizer
tokenizer = tiktoken.get_encoding("gpt2")

# Encode a single token (must be an **existing** token)
word = "hyperrealistic"
token_id = tokenizer.encode(word)[0]
print("Token ID:", token_id)

print(
    f"Parts of the word: {[tokenizer.decode([t]) for t in tokenizer.encode(word)]}"
)
```

```python
import tiktoken

# Load the GPT-2 tokenizer
tokenizer = tiktoken.get_encoding("gpt2")

# Encode a single token (must be an **existing** token)
token_id = tokenizer.encode_single_token("hyper")
print("Token ID:", token_id)
```

```python
import tiktoken

tokenizer = tiktoken.get_encoding("gpt2")

text = "The hyperrealistic painting was stunning."

tokens = tokenizer.encode(text)
print("Token IDs:", tokens)
print("Subword Tokens:", [tokenizer.decode([t]) for t in tokens])
```

```python
import tiktoken

tokenizer = tiktoken.get_encoding("gpt2")

text = "The hyperrealistic painting was stunning."

tokens = tokenizer.encode(text)
decoded_tokens = [tokenizer.decode([t]) for t in tokens]

print("Tokens:", tokens)
print("Decoded Token Parts:", decoded_tokens)
print("Reconstructed Text:", tokenizer.decode(tokens))
```
